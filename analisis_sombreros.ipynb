## castear la variable fecha que esta tipo float a string y quedarnos solo con los primero 4 digitos

```python
fecha_string = df_final['fecha'].astype(str).str[:4]
df_final['fecha'] = fecha_string
df_final['fecha']
# meses = {
#     'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04',
#     'mayo': '05', 'junio': '06', 'julio': '07', 'agosto': '08',
#     'septiembre': '09', 'octubre': '10', 'noviembre': '11', 'diciembre': '12'
# }

# meses_numerico_string = df_final['mes'].str.lower().map(meses)
# df_final['mes'] = meses_numerico_string
# df_final[['fecha', 'mes']]

```

## La variable **mes** esta por nombre, entonces atraves de un diccionario python mapeamos el mes de nombre -> número del mes
* como estan en mayuscula lo idel siempre es pasar a minuscula todos los valores de la variable **mes**

```python
meses = {
    'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04',
    'mayo': '05', 'junio': '06', 'julio': '07', 'agosto': '08',
    'septiembre': '09', 'octubre': '10', 'noviembre': '11', 'diciembre': '12'
}

meses_numerico_string = df_final['mes'].str.lower().map(meses)
df_final['mes'] = meses_numerico_string
df_final[['fecha', 'mes']]
```

## Ahora que las 2 variables (fecha,mes) son string las puedo concatenar para tenerla en formato YYYY-MM y asi poderla castear al tipo **Date** y poder hacer operaciones si es necesario


```python

fecha_completa_string = df_final['fecha'] + '-' + df_final['mes']
fecha_completa_string
```

## Casteo fecha_complet_date al tipo **Date** y se la asigno a la variable fecha

```python
fecha_completa_date = pd.to_datetime(fecha_completa_string, format='%Y-%m', errors='coerce', dayfirst=False).dt.to_period('M')
df_final['fecha'] = fecha_completa_date
df_final['fecha']
```

## Elimino la columna mes, por que no es necesaria
* inplace=True sirve para que se fuerze el guardado, si no se coloca al querer imprimir **df_final** en celdas posteriores te vuelve aparecer la columna **mes**\n
## Nota: puedes probar quitandolo el parametro completo o colocando en False

```python
if 'mes' in df_final.columns:
	df_final.drop(columns=['mes'], inplace=True)
df_final
```

```python
cantidad_vendida_int = df_final['cantidad_vendida'].astype(int)
df_final['cantidad_vendida'] = cantidad_vendida_int
df_final
```

## Encontramos que la variable **costo_envio** tiene muchos null, entonces vamos a observar a traves de un grafico de bigote que tan dispersos estan los valores

```python
plt.boxplot(df_final['costo_envio'].dropna())  # dropna() para evitar problemas con valores NaN
plt.title('Boxplot de Costo Envio')
plt.ylabel('Costo Envio')

# Mostrar el gráfico
plt.show()
```

## Se encontraron outliers, entonces podemos importar por la median() del metodo pandas

```python
median_costo_envio = df_final['costo_envio'].median()
df_final['costo_envio'] = df_final['costo_envio'].fillna(median_costo_envio)
df_final['total_facturado'] = df_final['costo_envio'] + df_final['total_venta']
df_final[['total_venta','costo_envio', 'total_facturado']]

```

## Pasamos a minuscula los valores de la variable **país**

```python
df_final['país'] = df_final['país'].str.lower().str.strip()
df_final['país']
```

## Agrupamos por pais y sumamos el total facturado por pais

```python
total_facturado_pais =  df_final.groupby('país').agg({
    'total_facturado': 'sum',
})
total_facturado_pais
```

# Graficamos Total facturado por **País**

```python
import matplotlib.ticker as mticker

ax = total_facturado_pais.plot.bar(figsize=(15, 8))
for p in ax.patches:
    ax.annotate(f'${p.get_height()/1000:.1f}K', (p.get_x() * 1.005, p.get_height() * 1.005))
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: f'{x/1000:,.0f}K'))
title_font = {
    'size': 16,
    # 'name': 'Helvetica'
}
ax.set_title('Total Facturado Por País', fontdict=title_font)
ax.set_xlabel('País')
ax.set_ylabel('Total en miles')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Total facturado por **AÑOS**

```python
import matplotlib.pyplot as plt
import plotly.express as pio



df_emiratos = df_final[df_final['país'] == 'emiratos arabes']
ventas_por_fecha = df_emiratos.groupby('fecha')['total_facturado'].sum()
# Convertir el índice a cadenas
ventas_por_fecha.index = ventas_por_fecha.index.astype(str)

plt.figure(figsize=(10, 6))
plt.plot(ventas_por_fecha.index, ventas_por_fecha.values, marker='o', linestyle='-', color='b')

plt.title('Ventas Totales en Emiratos Árabes a través de los ultimo 4 años', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Total Facturado (en dólares)', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.grid(True)
plt.tight_layout()
plt.show()
```
